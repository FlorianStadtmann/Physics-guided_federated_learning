```python
import numpy as np
from matplotlib import pyplot as plt
import torch
```

```python
def ELU(x):
    return np.where(x<=0,0.5*(np.exp(x)-1),x)
```

```python
def ReLU(x):
    return x*(x>0)
```

```python
def GELU(x):
    return 0.5*x*(1+np.tanh(np.sqrt(2/np.pi)*(x+0.044715*x**3)))
```

```python
def PReLU(x):
    return np.where(x>0,x,0.1*x)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.arange(-6,3,0.001)

plt.rcParams["figure.figsize"] = [5,2.5]

relu = ReLU(x)
prelu = PReLU(x)
elu = ELU(x)
gelu = GELU(x)

plt.plot(x,relu)
plt.plot(x,prelu)
plt.plot(x,elu)
plt.plot(x,gelu)

plt.legend(["ReLU(x)","PReLU(x)","ELU(x)","GELU(x)"],loc=2)

plt.yticks(ticks=[-1,0,1,2])

plt.xlim([-5,2])
plt.ylim([-0.7,2])

plt.xlabel("x")
plt.ylabel("f(x)")

plt.grid(alpha=0.5)
plt.savefig(format="pdf",fname="activationfunctions.pdf")
```

```python
x = np.arange(-10,10,0.001)

plt.rcParams["figure.figsize"] = [5,2]

sig = sigmoid(x)
tan = tanh(x)

plt.plot(x,sig)
plt.plot(x,tan)

plt.legend(["Logistic(x)","Tanh(x)"],loc=2)

#plt.yticks(ticks=[-1,0,1,2])

plt.xlim([-5,5])
plt.ylim([-1.5,1.5])

plt.xlabel("x")
plt.ylabel("f(x)")

plt.grid(alpha=0.5)
plt.savefig(format="pdf",fname="sigmoidal.pdf")
```
